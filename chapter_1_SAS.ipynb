```sas
data fig1_1;
    do t=1 to 120;
       z=5000+20*rannor(1234);  
       output;
    end;  run;
    
data fig1_1;
    set fig1_1;
    date=intnx('month', '1jan80'd, _n_-1);
    format date monyy.; run; 
    
proc sgplot;
   	series x=date y=z; run; 
```

```sas
data fig1_2;
    do t=1 to 120;
        x=0.5*t;
        z=0.5*t + rannor(1234);
        output;
    end; run;
    
data fig1_2; set fig1_2;
    date = intnx('month', '1jan80'd, _n_-1);
    format date monyy.; run;
    
proc sgplot data=fig1_2;
    series x=date y=x;
    series x=date y=z;
    yaxis label='z';
run;
```
